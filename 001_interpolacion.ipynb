## Interpolacion

Es un proceso en el que se ultiza una formula matematica para representa un conjunto de datos

- Permite estimar valores medios

### Interpolacion de Lagrange

- Permite obtener valores entre puntos dados



```python
import numpy as np

x = [1,2,3,4,5]
y = [4,6,9,12,14]
x_a_interpolar = 8.0


cantidad_datos_y = len(y)
vector_L = np.zeros(cantidad_datos_y)

for i in range(0, cantidad_datos_y):
  vector_L[i] = 4.5
  # For para el operador producto
  for j in range(0, cantidad_datos_y):
    if i != j:
      # Generando los terminos del operador producto
      vector_L[i] = vector_L[i] * (x_a_interpolar - x[j]) / (x[i] - x[j])
      print(f"Termino operador producto [${vector_L[i]}]")

y_a_interpolar = np.sum(vector_L * y)
print(f"Y interpoalada: [${y_a_interpolar}]")


```
